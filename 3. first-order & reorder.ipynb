## 1. 첫 주문인지 확인 하기

```python
sales['결제일']=pd.to_datetime(sales['결제일'],infer_datetime_format=True)
day_group=sales.groupby('아이디')['결제일']
```

```python
sales['첫 주문일']=day_group.transform(np.min)
sales.loc[sales['결제일'] == sales['첫 주문일'],'첫주문여부'] = True
sales.loc[sales['결제일'] != sales['첫 주문일'],'첫주문여부'] = False
```

```python
sales.head()
```

### 1) 특정 고객의 첫 주문일과 재구매 확인

- 해당 고객은 1월 22일 첫 구매를 했으며, 24번의 재구매를 했음

```python
sales[sales['아이디'] =='nha43']
```

### 2) 월 별 첫 주문 고객 수

```python
sales['결제월']=sales['결제일'].dt.month
```

```python
sales_table=sales.pivot_table(index='결제월',columns='첫주문여부',values='아이디',aggfunc='count',margins=True)
sales_table
```

첫 주문 고객의 비율

```python
table2 = sales_table.div( sales_table.iloc[:,-1], axis=0 ).round(2)
table2
```

- 1월에는 첫구매자 비율이 73%, 5355명으로 가장 많았고, 점차 재구매자가 증가해서 5월에는 재구매자 비율이 73%로 역전
- 매달 약 3,000명의 첫 구매 회원이 유입되고 있으며, 재구매자가 증가하여 전체적인 주문 볼륨이 늘어나고 있음

# 2. 재구매까지 소요되는 일자 보기

```python
s_reorder=sales[(sales['첫주문여부']==False)]
```

```python
s_reorder.head()
```

```python
day_group2=s_reorder.groupby('아이디')['결제일']
s_reorder['두번째 주문일']=day_group2.transform(np.min)
s_reorder.loc[s_reorder['결제일'] == s_reorder['두번째 주문일'],'두번째 주문여부'] = True
s_reorder.loc[s_reorder['결제일'] != s_reorder['두번째 주문일'],'두번째 주문여부'] = False
```

```python
s_reorder=s_reorder[['주문번호','두번째 주문일','두번째 주문여부']]
```

```python
sales=sales.merge(s_reorder,on='주문번호',how='left')
sales
```

[정리]

1) 두번째 주문일이 있으면서, 두번째 주문여부가 True 인 주문이 2번째 구매

2) 두번째 주문일이 있으면서, 두번째 주문여부가 False 인 주문은 2번째 이후 재구매

3) 두번째 주문일이 없는 주문일 경우, 첫구매 이후 재구매가 이루어지지 않은 고객 

```python
sales['재구매 소요기간'] = sales['두번째 주문일'] - sales['첫 주문일']
sales
```

```python
sales['재구매 소요시간']=(sales['재구매 소요기간'].dt.total_seconds() / 3600).round(0)
```
